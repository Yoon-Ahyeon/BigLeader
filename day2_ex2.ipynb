### "riss.kr" 홈페이지를 사용하여 학위 논문 제목 끌고오기

```python
from selenium import webdriver
# import chromedriver_autoinstaller as ca
from selenium.webdriver.common.by import By
import time, os, sys

#크롬 브라우저를 제어하는 driver 변수
driver = webdriver.Chrome()

#웹사이트 오픈하기
driver.get('https://www.riss.kr')
#윈도우 창으로 최대로 띄우기
#반응형 웹사이트 경우, element 위치 같은 것들이 변경되기 쉽기 때문에 기본적인 사이트를 맞춰놓고 시작한다. 
driver.maximize_window()

#새 창(팝업창) 제어하기
main = driver.window_handles #열린 창의 개수 저장

#첫번째로 뜨는 창은 MAIN창
for handle in main:
    if handle != main[0] :
        driver.switch_to.window(handle) #저장한 개수만큼 창을 이동한다. (다음 창으로 이동한다?)
        time.sleep(1)
        driver.close()

driver.switch_to.window(main[0])

#검색창에 검색어 넣기

time.sleep(2)
#Id=query(검색창)에 빅데이터라고 입력한다. (개행문자 '\n'은 엔터키 입력을 의미한다.)
driver.find_element(By.ID, 'query').send_keys('빅데이터'+ '\n') 
# driver.find_element(By.ID, 'query').clear() - 검색창을 지워주는 함수

time.sleep(2)
#Id와 Name이 없고 text만 존재한다. ; Link Text (웹페이지 어디있든 간에 클릭해준다.)
# driver.find_element(By.LINK_TEXT, '국내학술논문').click()

#Link_Text가 존재하지 않을 때 사용가능한 만능키 !!! = xPath (절대경로 or 상대경로) - 웹사이트에서 가장 유일한(근접한) 값을 찾는다. 
driver.find_element(By.XPATH, '//*[@id="tabMenu"]/ul/li/div/ul/li[3]/a').click()


'''                                              여기까지가 저번에 학습한 내용                                                     '''

#학위논문 제목 파싱하기

from bs4 import BeautifulSoup

#현재 타켓 페이지 전체를 가져온다. 
soup = BeautifulSoup(driver.page_source, 'html.parser')

#5초 후 웹사이트 닫기
time.sleep(5)
driver.close()
```

```python
#<p class = "title">인 내용 추출하기 
soup.find_all('p', class_ = 'title')[0].get_text()
```

```python
for i in soup.find_all('p', class_ = 'title'):
    print(i.get_text())
```

*class에서 title 제목을 통해서 그 안에 들어가 있는 정보들을 추출한다. 

class는 예약어이기 때문에 class 뒤에 언더바를 붙여 'class_'로 작성하거나 생략한다.
생략은 알아보기 힘드니 class_를 붙여주는 것을 추천한다. 

```python
#게시물 제목 추출하기 
title = soup.find('div', class_ = 'srchResultListW').find_all('p', class_ = 'title')
```

```python
for i in title:
    print(i.get_text())
```

```python
for idx, i in enumerate(title, start = 1): 
    print(idx, i.get_text())
```

```python
thesis = soup.find('div', class_ = 'srchResultListW')

title = thesis.find_all('p', class_ = 'title')
author = thesis.find_all('span', class_ = 'writer')
company = thesis.find_all('span', class_= 'assigned')
```

```python
for idx, i in enumerate(title, start = 1): 
    print(idx, i.get_text())
```

```python
for idx, i in enumerate(author, start = 1): 
    print(idx, i.get_text())
```

```python
for idx, i in enumerate(company, start = 1): 
    print(idx, i.get_text())
```

```python
#출력을 파일로 저장하기 

import sys
f_name = 'c:/py_temp/riss.txt'

#현재의 표준 출력 방향 저장  (f_origin 변수에 저장하겠다.)
f_origin = sys.stdout 

#file을 오픈하는 과정을 'file'이라는 별칭을 사용하겠다. 
#'a'는 기존 내용을 저장하면서 새로운 내용을 추가하겠다. (파일 내용 유지)
#'w'는 기존에 있던 내용들을 다 지우고 새로 작성하겠다. (파일 내용 삭제)
with open(f_name, 'w', encoding='UTF-8') as file: 

    #표준 출력 방향 - 지정된 파일로 텍스트를 입력하겠다. 
    sys.stdout = file

    for i in title:
        print(i.get_text().strip())

#출력된 내용을 변수에 저장하겠다. 
sys.stdout = f_origin

print('데이터 수집을 완료했습니다.')
print('수집 결과는 %s에 저장했습니다.'%f_name)
```
