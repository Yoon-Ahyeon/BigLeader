#### tf * idf

TFIDF

tf(d, t) : 문서에서 단어의 등장 횟수
(단어가 여러 번 발생할수록 값이 증가한다.)

idf(d, t) : 단어가 흔하게 문서에 등장할 수록 값이 감소하며, 단어가 여러 번 발생하지 않을 때는 증가한다.

<!-- tf - idf → 비례관계

- tf가 클 수록 idf도 크다
  = 단어 빈도수가 높다 -->

- df(t)가 크면, 단어가 흔하게 문서에 등장한다.

즉, 문서당 빈도수가 적고 흔하지 않은 단어들을 추출하기 위해 사용한다.

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm #선형함수들이 정의되어 있다.
```

```python
#document 각 열은 영화 대사에 등장한 단어(사랑, 돈, 운명, 증오)의 빈도
document1 = np.array([1, 1, 1, 0])
document2 = np.array([0, 1, 1, 1])
document3 = np.array([0, 5, 5, 5])
```

#### document 사이 유클리디안 거리 구하기

d2를 기준으로 했을 경우,
유클리디안 거리는 document1 + documnet2 > document2 + document3 이기 때문에
document1과 document2의 유사도가 더 크다.

#### cosine 유사도 구하기

```python
#cosine 유사도 구하기 (documnet1과 document2 사이의 유사도)
dot(document1, document2) / (norm(document1)*norm(document2))
```

```python
#cosine 유사도 구하기 (documnet2과 document3 사이의 유사도)
dot(document2, document3) / (norm(document2)*norm(document3))
```

```python
#cosine 유사도 구하기 (documnet1과 document3 사이의 유사도)
dot(document1, document3) / (norm(document1)*norm(document3))
```

#### 자카드 유사도

자카드 유사도(d1, d2) = (d1 교집합 d2) / (d1 합집합 d2)
</br></br>
자카드 유사도가 1인 경우?

d1 교집합 d2 = d1 합집합 d2

→ d1, d2가 동일하다.

```python
#불용어 제거
tfidf = TfidfVectorizer(stop_words = 'english')
```

```python
tfidf_m = tfidf.fit_transform(df['overview'])
tfidf_m.shape
#단어 32,350개, overview 10,000개
```

```python
tfidf_m #3억개의 요소 .. (영화끼리 유사도)
```

```python
#코사인 유사도
cos_sim = cosine_similarity(tfidf_m, tfidf_m)
```

```python
data = dict(zip(df['title'], df.index)) #인덱스와 영화 제목을 딕셔너리 형태로 생성
i = data['Toy Story']
print(i)
```

```python
def recommend(t, cosine_sim = cos_sim): #t에는 제목 전달
  idx = data[t] #딕셔너리 키 전달 - value(인덱스 추출)
  #idx 해당하는 cosine 유사도를 가져온다.
  #print(cosine_sim[idx])
  sim_scores = list(enumerate(cosine_sim[idx])) #인덱스번호와 유사도를 같이 내보내기 위해 열거형 사용
  print(sim_scores)
  ss = sorted(sim_scores, key = lambda x:x[1], reverse = True) #내림차순 정렬
  #print(ss)
  ss_scores = ss[1:11] #1~10등까지 출력
  res = [i[0]  for i in ss_scores]
  return df['title'].iloc[res] #영화제목 출력을 하기 위해서 iloc를 넣는다?? [iloc 찾아보자..]
```

```python
recommend('Toy Story')
```
