### 43.2 범위 판단하기

 [ ] (대괄호) 안에 숫자 범위를 넣으며 * 또는 +를 붙인다.

 숫자 범위는 0-9처럼 표현하며 *는 문자(숫자)가 0개 이상 있는지, +는 1개 이상 있는지 판단한다.

```python
re.match('[0-9]*', '1234') #0개 이상 있는가?
```

```python
re.match('[0-9]+', '1234') #1개 이상 있는가?
```

```python
re.match('[0-9]+', 'abcd')
```

```python
re.match('[0-9]*', 'abcd')
```

```python
re.match('a*b', 'b') #b에 a가 0개 이상 있는가?
```

```python
re.match('a+b', 'b') #b에는 a가 1개 이상 있는가?
```

```python
re.match('a*b', 'aab') #aab에는 a가 0개 이상 있는가?
```

```python
re.match('a+b', 'aab') #aab에는 a가 1개 이상 있는가?
```

?는 ? 앞의 문자(범위)가 0개 또는 1개인지 판단

 .은 .이 있는 위치에 아무 문자(숫자)가 1개 있는지 판단

```python
#문자가 한개 있는지 판단하기

re.match('abc?d', 'abd')
```

```python
re.match('ab[0-9]?c', 'ab3c')
```

```python
re.match('ab.d', 'abxd') #. 위치에 다른 숫자 하나가 있는가?
```

문자 뒤에 {개수} 형식을 지정

문자열의 경우에는 문자열을 괄호로 묶고 뒤에 {개수} 형식을 지정

```python
re.match('h{3}', 'hhhello')
```

```python
re.match('(hello){3}', 'hellohellohelloworld')
```

```python
re.match('[0-9]{3}-[0-9]{4}-[0-9]{4}', '010-1000-1000')
```

```python
re.match('[0-9]{3}-[0-9]{4}-[0-9]{4}', '010-1000-100')
```

```python
#일반 전화 번호 형식에 맞는지 판단
re.match('[0-9]{2, 3}-[0-9]{3, 4}-[0-9]{4}', '02-100-1000')
```

숫자와 영문 문자를 조합해서 판단

```python
re.match('[a-zA-Z0-9]+', 'Hello1234') #a-z 1개, A-Z 1개, 0-9 1개 이상이면 매칭완료
```

```python
re.match('[A-Z0-9]+', 'hello1A3') #A-Z, 0-9까지 숫자 중 하나 이상이면 매칭 완료
```

```python
re.match('[가-힣]+', '홍길동')
```

문자(숫자) 범위 앞에 ^를 붙이면 해당 범위를 제외

```python
re.match('[^A-Z]+', 'Hello') #대문자 제외
```

```python
re.match('[^A-Z]+', 'hello')
```

```python
re.search('^[A-Z]+', 'Hello') #대문자로 시작하면 패턴에 매칭
```

```python
re.search('[0-9]+$', 'Hello1234') #숫자로 끝나면 패턴에 매칭
```

특수 문자를 판단할 때는 특수 문자 앞에 \를 붙인다.

```python
re.search('\*+', '1 ** 2')
```

```python
re.match('[$()a-zA-Z0-9]', '$(document)') #괄호 안에 a-z, A-Z, 0-9 중 한 개 이상 존재
```

```python
re.match('\d+', '1234')
```

```python
re.match('\D+', 'Hello') #숫자를 제외한 모든 문자
```

```python
re.match('\w', 'Hello_1234') #영어 대소문자, 숫자, 밑줄 문자 중 한 개 이상
```

```python
re.match('\w+', '(:)')
```

공백은 ' '처럼 공백 문자를 넣어도 되고, \s 또는 \S로 표현

```python
re.match('[a-zA-Z0-9 ]+', 'Hello 1234')
```

### 43.4 문자열 바꾸기

```python
re.sub('apple|orange', 'fruit', 'apple box orange tree')
```

```python
re.sub('[0-9]+', 'n', '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
def multiple10(m):
  n = int(m.group())
  return str(n*10)

re.sub('[0-9]+', multiple10, '1, 2, Fizz 4 Buzz Fizz 7 8')
```

```python
re.sub('[0-9]+', lambda m:str(int(m.group()) * 10), '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
re.sub('([a-z]+) ([0-9]+)', '\\2 \\1 \\2 \\1', 'hello 1234')  #그룹 2 1 2 1 순으로 변경
```

```python
re.sub('({\s*)"(\w+)":\s*"(\w+)"(\s*})', '<\\2>\\3</\\2>', '{ "name": "james" }')
```

```python
re.sub('({\s*)"(?P<key>\w+)":\s*"(?P<value>\w+)"(\s*})',
       '<\\g<key>>\\g<value></\\g<key>>',
       '{ "name": "james" }')
```

### 43.5 연습문제: 이메일 주소 검사하기

```python
import re

p = re.compile(r'([a-zA-Z0-9+-._]+)@([a-z-]+)\.([a-z]+)')
emails = ['python@mail.example.com', 'python+kr@example.com',              # 올바른 형식
          'python-dojang@example.co.kr', 'python_10@example.info',         # 올바른 형식
          'python.dojang@e-xample.com',                                    # 올바른 형식
          '@example.com', 'python@example', 'python@example-com']          # 잘못된 형식

for email in emails:
    print(p.match(email) != None, end=' ')
```

#### 범위 판단하기

[ ] (대괄호)는 한 문자를 나타내는 것

ex) [a] : 문자 a를 의미,

[ab] : 문자 a, b 중 하나를 의미,

[a-c] : 문자 a, b, c 중 하나를 의미한다.

```python
#0-9까지 숫자 중 문자열에 하나라도 있으면 match한다.
re.match('[0-9]*', '1234')
```

```python
re.match('[0-5]*', '1234')
```

```python
# *은 0개 이상이기 때문에 match가 된 것
re.match('[5-9]*', '1234')
```

```python
re.match('[a-z]*', 'abc1234')
```

```python
re.match('[a-z0-9]*', 'abc1234')
```

```python
re.match('[0-9a-zA-Z가-힣ㄱ-ㅎ]*', 'ㅋㅋㅋ큭큵1234')
```

```python
res = re.match('[0-3]*', '1234')
res
```

```python
res_2 = re.match('[가-마]+', '1234') #매치가 되지 않으면 False로 처리된다.
res_2
```

```python
#변수에 값이 들어가있으면, True로 설정된다.
if res_2:
  print("매치가 됐어요 !")
else:
  print("매치가 안 됐어요 !")
```
